```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('serbian-rw.npz')
diff = data['diff']
W0 = data['W0']
O = data['O']
W = data['W']
```

```python
n = diff.shape[0]
m = diff.shape[1]
d, w, o, w0 = [], [], [], []
for i in xrange(n):
    for j in xrange(m):
        if O[i,j]:
            d.append(diff[i,j])
            w0.append(W0[i,j])
            o.append(O[i,j])
            w.append(W[i,j])
diff = np.array(d)            
W = np.array(w)
O = np.array(o)
W0 = np.array(w0)
```

```python
(diff/W0).mean()
```

```python
plt.axis([0,1,0,1])
plt.scatter(W0,W)
```

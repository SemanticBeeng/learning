```python
cards = [ztnbinom.rvs(3,.6) for i in xrange(10000)]
data = np.zeros((max(cards)))
for i in cards:
    data[i-1] += 1
data
```

```python
data = pd.DataFrame(data,columns=['Frequency'],index=range(1,len(data)+1))
data['Cues'] = [range(1,i+1) + ['exactly%d'%i] for i in data.index]
data['Number'] = data.index
data['Outcomes'] = 'plural'
data['Outcomes'][1] = 'singular'
```

```python
%%px
import pandas as pd
import numpy as np

import sys
sys.path.append('/home1/malouf/learning')
import ndl
```

```python
%%px --local
def activation(data,W):
    return pd.DataFrame([ndl.activation(c,W) for c in data.Cues],index=data.index)
```

```python
%%px --local
def accuracy(data, M):
    W = ndl.rw(data, M=M)
    A = activation(data, W)
    return np.mean(A.idxmax(1) == data['Outcomes'])
```

```python
%%px --local
def population_accuracy(M, pop=100):
    return np.mean([accuracy(data, M=M) == 1 for i in xrange(pop)])
```

```python
MAX_TRIALS = 250
P = {}
```

```python
def result(P,exp):
    plt.plot(range(1,len(P[exp])+1), P[exp], '-', linewidth=2)
    plt.title(exp)
    plt.xlabel('Trial Number')
    #plt.yscale('log')
    plt.suptitle('Proportion of 100 learners who label all 15 items correctly')
```

```python
%%time
dview.push(dict(data=data))
P['sg / pl'] = lview.map(population_accuracy,xrange(1,MAX_TRIALS))
```

```python
result(P,'sg / pl')
```
